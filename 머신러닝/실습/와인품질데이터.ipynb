```python
import pandas as pd
```

```python
redwine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';', header=0)
redwine["type"] = 'red'
redwine.head()
```

```python
whitewine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep=';', header=0)
whitewine["type"] = 'white'
whitewine.head()
```

```python
wine = redwine.append(whitewine)
wine.shape
```

```python
wine.columns = wine.columns.str.replace(" ", "_")
wine.head()
```

```python
# describe 요약통계(기수통계량) : 연속형 데이터의 기술통계량 확인 가능, 통계량은 변량의 측정값
wine.describe()
```

```python
#개별 통계량 확인 가능, 시리즈 로 확인 가능
wine.quality.describe()
```

```python
# 와인 품질의 1~10점중 유일한 값을 확인
sorted(wine.quality.unique())
```

```python
# 각 품질점수별 빈도 확인은 value_counts() # 내림차순은 ascending=False
wine.quality.value_counts().sort_index(ascending=False)
```

```python
# 와인 종류별 품질의 기술통계량 확인
wine.groupby("type")['quality'].describe()
```

```python
# 와인 종류별 품질의 사분위수   # 사분위수를 주기위한 범위 설정,  DataFrame으로 변경하고 type을 컬럼으로 설정
wine.groupby("type")['quality'].quantile([0, 0.25, 0.5, 0.75, 1]).unstack("type")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 와인 종류별 품질의 분포
red_q = wine.loc[wine['type'] == 'red', 'quality']
white_q = wine.loc[wine['type'] == 'white', 'quality']
```

```python
sns.set_style("darkgrid")
sns.distplot(red_q, norm_hist=True, kde=False, color='red', label='Red Wine')
#   분포 플롯
sns.distplot(white_q, norm_hist=True, kde=False, color='blue', label='White Wine')
plt.title("Distribution of Quality of Wine Type")
plt.xlabel("Quality Score")
plt.ylabel("Density")
plt.legend();
# quality score 가 5~7에 걸쳐 있음
```

```python
# 와인 종류별 품질의 차이의 통계적 유의성 검정; 귀무가설 정의 > 표본 추출 > 통계적유의성 검정 > 결과에 따라 귀무가설 T F 결정
# 2개의 모집단(화이트, 레드)의 평균의 차이가 있는지 검정
wine.groupby("type")["quality"].aggregate(["std", "mean"])
```

## 귀무가설 : 와인 종류별 품질 평균 차이가 없음

- p-value < 0.05 : 평균차이 있음
- p-value > 0.05 : 평균차이 없음


```python
import statsmodels.api as sm

t_stat, p_value, df = sm.stats.ttest_ind(red_q, white_q)
                    # stats 의 ttest_ind로 두개 비교해서 t_stat, p_vale, df 밀도함수그릴떄 기울기를 받을 수 있음
"t-stat : {:.3f}, p-value : {:.4f}".format(t_stat, p_value)
```

```python
# p_value값이 0.05보다 작아 평균의 차이가 있음, 종류별 품질 차이가 있음
```

```python
wine_corr = wine.corr()
wine_corr
```

```python
# 양의 상관관계 : 변수x가 커지면 y도 커짐
# 음의 상관관계 : 변수x가 커지면 y가 작아짐
```

```python
#양의 상관관계
wine_corr.loc[wine_corr["quality"]>0,"quality"]
# 와인상관관계의. 와인 퀄리티의 상관계수가 0보다큰 것을 꺼내라, 퀄리티에 관한 부분만
```

```python
# 양의 상관관에서 가장큰 것은 알콜 도수
```

```python
# 음의 상관관계
wine_corr.loc[wine_corr['quality'] < 0, 'quality']
```

```python
# 음의 상관관계에서 가장 큰 것은 밀도(density)
```

## 산점도 행렬

```python
import numpy as np
```

```python
red_sample = wine.loc[wine.type == 'red', :]
white_sample = wine.loc[wine.type == 'white', :]
```

```python
red_idx =  np.random.choice(red_sample.index, replace=True, size=200)
white_idx = np.random.choice(white_sample.index, replace=True, size=200)
```

```python
wine_sample = red_sample.loc[red_idx, ].append(white_sample.loc[white_idx, ])
wine_sample.head()
```

```python
wine_sample.info()
```

```python
sns.set_style("dark")
sns.pairplot(wine_sample, vars=["quality", "alcohol", "residual_sugar"], 
            kind="reg", plot_kws = {"ci:": False, "x_jitter": 0.25, "y_jitter": 0.25}, 
            diag_kind="hist", diag_kws={"bins": 10, "alpha": 1},
            hue="type", palette=dict(red="red", white="blue", markers=["o", "s"]));
```
